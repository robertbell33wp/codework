```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import numpy as np
```

```python
df_time = pd.read_csv('ConstructionTimeSeriesDatav2.csv')
```

```python
fig,ax = plt.subplots()

'''Plot 2 Time Series'''
ax.plot(df_time['Month-Year'],df_time['Private Construction'],color='b',linewidth=.5)
ax.plot(df_time['Month-Year'],df_time['Public Construction'],color='r',linewidth=.5)

'''Remove Spines'''
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

'''Set White Space at Top and Bottom'''
ax.set_ylim(0,90000)

'''Set Tick Lmits'''
ax.xaxis.set_tick_params(which = 'both', top = False, bottom = True, labelbottom = True, direction='in')
ax.yaxis.set_tick_params(which = 'both', right = False, left = True, labelleft = True, direction='in')
ax.set_xticks(df_time['Month-Year'][::9])
ax.set_xticklabels(df_time['Month-Year'][::9])

'''Set Axis Labels'''
ax.set_xlabel('Month-Year',fontsize='14',fontname='Sans')
ax.set_ylabel('Spending (in Millions of $)',fontsize='14',fontname='Sans')

'''Add Legend'''
ax.legend(loc=1)

'''Set Figure Size'''
fig.set_size_inches(20,8)

'''Set Chart Title'''
fig.suptitle("U.S. Private vs. Public Construction (2002-2014) ",fontsize='20',fontname='Sans')

'''Add shaded area to highlight aspect'''
ax.axvspan(25,85,ymin=0,ymax = 300,facecolor='k',alpha=0.1)
ax.text(55,43000,'3.5x Spending Gap',fontsize=16,verticalalignment='center',
        horizontalalignment='center',fontname='Sans')


```

Between 2002 and 2014 the gap between Public and Private construction spending reached its highest levels beginning in February 2004 and ending around February 2008.  When focusing on residential construction, this gap may have impacted lower income households the most.  The increase of, more expensive, private residential constructions may have displaced millions of households who would have benefited from more public housing options
